```python
entrada = [12, 15, 18, 22, 25, 30, 35, 40, 45] #minha tentativa

def calcula_mediana(dados):
    dados.sort()
    posicao_1 = int((len(dados) / 2) - 1) #posição 1 (se for par)
    posicao_2 = int(len(dados) / 2) #posicao 2 (par ou impar)

    if len(dados) % 2 == 0: #se for par (número de itens)
        mediana = float((dados[posicao_1] + dados[posicao_2])/2) #média entre duas centrais
    else:
        mediana = float(dados[posicao_2]) #se for ímpar, é a posição central
    return mediana

calcula_mediana(entrada)
```

```python
#SOLUÇÃO DO GEMINI
def calcula_mediana(dados):
    dados_ordenados = sorted(dados) #organizou os dados
    n = len(dados_ordenados) #gerou n - que é o len da lista
    meio = n // 2 #// DIVISÃO INTEIRA - gerou meio (divisão inteira = posição do meio)
    
    if n % 2 != 0: #se for impar
        mediana = dados_ordenados[meio] #mediana = dado do meio
    else: #se for par
        mediana = (dados_ordenados[meio - 1] + dados_ordenados[meio]) / 2 #mediana = média entre (meio -1) e (meio)
        
    return float(mediana)

# Teste com os seus dados
descordenada = [-16, 48, -43, -13, -99, -26, 48, -11, 80, 20, 68, 24, -1]
print(calcula_mediana(descordenada)) # Retorno correto: -1.0
```

```python
#solução do gemini
dicionario = {"item": ["condicionador", "condicionador", "condicionador", "chocolate", "chocolate", "chocolate"], "quantidade": [1, 2, 3, 4, 5, 6]}

def calcula_media_item(dicionario):
    # Dicionários auxiliares para armazenar a soma e a contagem
    somas = {}
    contagens = {}
    
    # Extrai as listas do dicionário de entrada
    itens = dicionario['item']
    quantidades = dicionario['quantidade']
    
    # Itera sobre os itens e quantidades simultaneamente
    for item, qtd in zip(itens, quantidades):
        if item not in somas:
            somas[item] = 0
            contagens[item] = 0
        
        somas[item] += qtd
        contagens[item] += 1
    
    # Calcula a média final para cada item
    resultado = {}
    for item in somas:
        resultado[item] = somas[item] / contagens[item]
        
    return resultado
calcula_media_item(dicionario)
```

```python
import pandas as pd #Solução usando PANDAS (maravilhoso)

def calcula_media_item(dicionario):
    df = pd.DataFrame(dicionario)
    # Agrupa por item e calcula a média da quantidade
    media_df = df.groupby('item')['quantidade'].mean()
    # Converte de volta para o formato de dicionário solicitado
    return media_df.to_dict()

calcula_media_item(dicionario)
```
