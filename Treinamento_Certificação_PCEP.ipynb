# 3) Dicionários

```python
#Transformar LISTA em DICIONARIO
dicionario = {} #abre dicionário

lista = [10,20,30,40] #dá lista
for i in lista: #para cada item
  dicionario[i] = i**2 #o iterado recebe i**2
dicionario #print
```

```python
#Transformar Dicionário em LISTA
#list comprehension, for ou list()
keys = [chave for chave in dicionario]
print(keys)

values = [value for value in dicionario.values()]
print(values)
```

```python
# 1. Transformando as chaves em lista - FOR
keys = []
for item in dicionario: #pode acessar o dicionário direto
    keys.append(item)

# 2. Transformando os valores em lista
values = []
for item in dicionario.values(): #tem que acessar o dicionario.values()
    values.append(item)

print(keys)
print(values)
```

# 5) Funções

```python
#estudar funções com if, sem else e com comando embaixo:
# Quando encontra RETURN, a função fecha (se não tiver recursividade)

def f(x=3): #ignorar x=3
  if x == 0:
    return 0 #encerrar (return 0) quando x == 0
  return x + f(x-1) #recursividade - vai executar a função dentro da função
print(f(3))
# f de 3 == retorna 3 e executa de novo (soma próximo)
# f de 2 == retorna 2 e executa de novo (soma próximo)
# f de 1 == retorna 1 e executa de novo (soma próximo)
# f de 0 == returna 0 (= break) (soma 0)
#resultado == 3+2+1 == 6
```
