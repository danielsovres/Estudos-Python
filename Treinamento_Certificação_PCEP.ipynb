```python
x = 10
while x > 5: #só funciona enquanto x é mais que 5
    x -= 3
    # primeira = x=10 , x>5, então 10-3 = 7 (imprime 7)
    # segunda = x=7, x>5, então 7-3 = 4 (imprime 4)
    print(x) #imprime 7 e 4
```

```python
lista = [1, 2, 3] #len = 3
for i in lista:
    lista.append(i) #vai adicionar um número a cada loop
    if len(lista) > 5: #se o len for maior que 5 (portanto, 6) = break
        break #então vai brekar quando a lista tiver 6 caracteres
print(lista)
```
