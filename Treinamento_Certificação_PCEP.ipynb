```python
#estudar funções com if, sem else e com comando embaixo:
# Quando encontra RETURN, a função fecha (se não tiver recursividade)

def f(x=3): #ignorar
  if x == 0:
    return 0
  return x + f(x-1) #recursividade - vai executar a função dentro da função
print(f(3))
# f de 3 == retorna 3 e executa de novo
# f de 2 == retorna 2 e executa de novo
# f de 1 == retorna 1 e executa de novo
# f de 0 == returna 0 (= break)
#resultado == 3+2+1 == 6
```
